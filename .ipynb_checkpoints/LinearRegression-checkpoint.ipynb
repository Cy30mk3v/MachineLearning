# Import library

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

# Loading data

```python
# height (cm)
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180, 183]]).T
# weight (kg)
y = np.array([[ 49, 50, 51,  54, 58, 59, 60, 62, 63, 64, 66, 67, 68]]).T
# Visualize data 
plt.plot(X, y, 'ro')
plt.axis([140, 190, 45, 75])
plt.xlabel('Height (cm)')
plt.ylabel('Weight (kg)')
plt.show()
```

# Using built-in library

```python
model_LR = LinearRegression().fit(X, y)
```

```python
y_pred = model_LR.predict(X)
```

```python
plt.scatter(X,y,color="red")
plt.plot(X,y_pred,color="blue",linewidth=3)
```

# Mathematical solution

With:
- $ \hat y = \bar x w $: The prediction result
- $ y $: The true result
<br> 
<br>
<b>We have:

- Error rate: $ e = (y - \hat y ) ^ 2 = (y - \bar x w)^2$ -> For each prediction (sample)
- Error prediction: $ \frac{1}{2}e = \frac{1}{2}(y - \hat y ) ^ 2 = \frac{1}{2}(y - \bar x w)^2 $ -> -> For each prediction (sample)
- Loss function: $ L = \frac{1}{2}\sum_{i=1}^N\ (y_i - \bar x_i w)^2 $ -> For whole prediction (whole dataset)
    <br> -> Loss function of Linear regression
- The problem of Linear regression -> Find $w$ to make $L$ as small as posible, a.k.a $ \underset{w}{\operatorname{argmin}}L(w) $ -> Find optimal point for function 
- If we put Y as an vector, we will have $ L= {\lVert y-\bar Xw \rVert}^2_2$ -> Total of square Euclidean norm distance from the prediction to the true result
